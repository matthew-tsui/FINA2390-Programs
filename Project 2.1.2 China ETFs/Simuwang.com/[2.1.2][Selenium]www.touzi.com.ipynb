```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import csv
import pandas as pd
import os
import time
import numpy as np
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

START_PAGE = 401
END_PAGE = 600
driver = webdriver.Firefox()
driver.get("https://www.touzi.com/simu/productlist-pn1.html#anchor")

#df_columns = ['ranking','fund_short_name','fund_company','investment_strategy','lastest_NAV','cumulative_return','recent_month_return','annalized_return','sharpe_ratio','year_rating']
#df_current_columns = ['Ranking','Fund Short Name','Investment Strategy','Fund Company','Lastest NAV','Cumulative Return']
df_etf = pd.DataFrame(columns=['Ranking','Fund Short Name','Investment Strategy','Dummy','Fund Company','Dummy','Recent Month Return','Dummy','Lastest NAV','Cumulative Return','Dummy','Dummy','Dummy','Annalized Return','Sharpe Ratio','Year Rating'])

def go_to_page(page):
    if page != 1:
        url = 'https://www.touzi.com/simu/productlist-pn'+ str(START_PAGE) + '.html#anchor'
        driver.get(url)
    else:
        return

def bypass_login():
    iclose = driver.find_element_by_xpath('/html/body/div[13]/a').click()
    print 'Bypass login success'
    print 'Ready to crawl pages...'

def main():
    go_to_page(START_PAGE)
    bypass_login()
    print 'Start at PAGE'+ str(START_PAGE)
    
    n = 1
    for CURRENT_PAGE in range(START_PAGE,END_PAGE+1):
        print '--------------------PAGE ' + str(CURRENT_PAGE) +' START--------------------'
        START_TIMESTAMP = time.time()
        for row in driver.find_elements_by_xpath('/html/body/div[7]/div[2]/table/tbody/tr'):
            #skip the first row
            if len(row.find_elements_by_tag_name("td")) == 0:
                continue
            #print 'Start handling '+ str(n) +'th row'
            for i in range(12):
                df_etf.loc[n, df_etf.columns[i]] = row.find_elements_by_tag_name("td")[i].text
            #print 'Finish handling '+ str(n) +'th row'
            #print '----------------------------------'
            n = n + 1
        print '--------------------PAGE ' + str(CURRENT_PAGE) +' FINISH--------------------'
        END_TIMESTAMP = time.time()
        print '-TIME USED:' + str(START_TIMESTAMP-END_TIMESTAMP) + '-'
        
        #store every first 10,20,30... pages to CSV for backup purpose
        if (CURRENT_PAGE % 50) == 0:
            filename = '[Bonus][touzi][Intrim Backup][' + str(CURRENT_PAGE) + ']Project 2.1.2 PAGE ' + str(START_PAGE) + 'to PAGE ' + str(CURRENT_PAGE) + 'Records.csv'
            df_etf.to_csv(filename, index=False,  encoding='utf-8')
            print '[Intrim Backup]Saved First '+ str(CURRENT_PAGE) + 'Pages Records to CSV'
    
        #click next page
        driver.find_element_by_xpath('/html/body/div[7]/div[2]/div/a[13]').click()
        time.sleep(3)
    
    filename = '[Bonus][touzi]Project 2.1.2 PAGE ' + str(START_PAGE) + 'to PAGE ' + str(END_PAGE) + 'Records.csv'
    df_etf.to_csv(filename, index=False, encoding='utf-8')
    print 'Exported to csv'
    driver.close()
        
    

if __name__ == "__main__":
    main()

```
